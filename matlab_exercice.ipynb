```python
import numpy as np, matplotlib.pyplot as plt, math
```

```python
X = np.arange(-(math.pi), math.pi, 0.01)
X
```

```python
y = np.sin(X)
```

```python
plt.plot(X, y, label='Sinus de pi')
plt.xlabel("Valeur de X")
plt.ylabel("sin(X)")
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
t = np.arange(0, (10 * math.pi), (math.pi / 50))
```

```python
plt.figure(figsize=(12, 8))
ax = plt.axes(projection='3d')
ax.plot3D(np.sin(t), np.cos(t), t)
ax.set_xlabel('sin(t)')
ax.set_ylabel('cos(t)')
ax.set_zlabel('t')
plt.show()
```
